# Modelos Relacionales

En bases de datos relacionales, un modelo relacional es una estructura de datos que representa la información almacenada en una base de datos. Un modelo relacional es una colección de tablas, donde cada tabla tiene un nombre y una serie de columnas. Cada fila de una tabla representa una entidad, y cada columna de una tabla representa un atributo de la entidad.

## Comparacion modelo relcional vs semiestructurado

| Modelo Relacional | Modelo Semiestructurado |
| --- | --- |
| Estructura fija | Estructura flexible |
| Tablas | Documentos, grafos, etc. |
| SQL | NoSQL |
| PostgreSQL, MySQL, SQLite | MongoDB, Cassandra, Redis |
| Normalización | Desnormalización |
| JOINs | No JOINs |
| Transacciones | No transacciones |
| ACID | BASE |
| Escalabilidad vertical | Escalabilidad horizontal |
| Esquema fijo | Esquema flexible |



Ambos:
* Proveen solución para almacenar datos

esquema

## Modelo relacional

• Relaciones: a cada tabla le llamamos relación
• Atributos: son las columnas de la relación
• Tuplas: son las filas de la relación


Podemos ver cada tabla (relacion) como una clase abstracta con sus atributos



Películas(id, nombre, año, categoría, calificación)

Donde las tuplas son cada instancia de la clase.

Cada atributo tiene un dominio, que es el conjunto de valores que puede tomar

Películas(id:int, nombre:string, año:int,
categoría:string, calificación:float)

Una instancia de un esquema es un conjunto de
tuplas para cada relación del esquema


* La instancia de un esquema sería un conjunto de instancias de clases
* Cada instancia de clase es una tupla
* Cada tupla tiene un valor para cada atributo
* Cada atributo tiene un dominio


El atributo másimportnate de cada relación se le denomina **llave**

en el caso anterior, la llave de la relación Películas es el atributo id

**lo subrayamos**

Películas(**id**:int, nombre:string, año:int, categoría:string, calificación:float)


## Terminología de LLaves

**Super llave (superkey)**: cualquier conjunto de atributos que determina a todo el resto

**Llave (candidata/minimal)**: cualquier conjunto de
atributos que determina a todo el resto, y ninguno de sus subconjuntos es una super llave

**Llave primaria**: una llave candidata que queremos
destacar (la subrayada en el esquema)




**Surrrogate key**: una llave generica que simplicada la busqueda de datos: id

**Llave foranea**: es una llave que hace referencia a otra tabla

Resumen SQL codigo:

```sql
CREATE TABLE actor (
    id SERIAL PRIMARY KEY,
    nombre VARCHAR(100)
);

CREATE TABLE pelicula (
    id SERIAL PRIMARY KEY,
    nombre VARCHAR(100)
);

CREATE TABLE actor_pelicula (
    id_actor INT,
    id_pelicula INT,
    FOREIGN KEY (id_actor) REFERENCES actor(id),
    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id)
);
```

Consulta 1: Liste el nombre de todos los actores

```sql
SELECT nombre FROM actores;
```

Consulta 2: Liste el nombre y la calificación de todas las películas

```sql
SELECT nombre, calificacion FROM peliculas;
```

Consulta 3: Liste el nombre y la calificación de todas las películas con
calificación inferior a 8.5

```sql
SELECT nombre, calificacion FROM peliculas WHERE calificacion < 8.5;
```

Consulta 4: Liste todas las películas de Nolan

```sql
SELECT nombre FROM peliculas WHERE director = 'C. Nolan';
```

Consulta 5: Liste todos los id de los actores de la película “Interstellar”

```sql
SELECT id_actor FROM actuo_en WHERE id_pelicula = 1;
```

Consulta 6: Liste cada actor junto a todas las películas en las que ha
actuado

```sql
SELECT actores.id as id, actores.nombre AS actor, peliculas.nombre AS pelicula
FROM actores
JOIN actuo_en ON actores.id = actuo_en.id_actor
JOIN peliculas ON actuo_en.id_pelicula = peliculas.id;
```

Consulta 7: Liste todas las películas en que actúe Leonardo DiCaprio y
que sean dirigidas por C. Nolan

```sql
SELECT actores.nombre AS nombre
FROM actores JOIN actuo_en ON actores.id = actuo_en.id_actor
JOIN peliculas ON actuo_en.id_pelicula = peliculas.id
WHERE actores.nombre = 'Leonardo DiCaprio' AND peliculas.director = 'C. Nolan';
``` 

Consulta 8: Liste todas las películas y la calificación en que actúe
Leonardo DiCaprio o que sean dirigidas por C. Nolan

```sql
SELECT peliculas.nombre AS nombre, peliculas.calificacion AS calificacion FROM peliculas JOIN actuo_en ON peliculas.id = actuo_en.id_pelicula JOIN actores ON actuo_en.id_actor = actores.id WHERE actores.nombre = 'Leonardo DiCaprio' OR peliculas.director = 'C. Nolan';
```

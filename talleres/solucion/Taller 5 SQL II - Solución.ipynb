#SQL
## Pregunta 1
Las películas de los 80s, ordenadas por calificación de mayor a menor.

```
%%sql
select *
from pelicula
where anho between 1980 and 1989
order by calificacion desc;
```

## Pregunta 2
Los nombres de los personajes que ha interpretado su actriz/actor favorito en los datos, ordenados por año.

```
%%sql
select personaje, p_anho
from actor inner join personaje on actor.nombre = personaje.a_nombre
where nombre ilike '%nicholson, jack%'
order by p_anho asc;
```

# Pregunta 3
Las películas en las que participó su actriz/actor favorito en los datos, ordenadas por calificación de mayor a menor.

```
%%sql
select pelicula.nombre, pelicula.calificacion
from actor inner join personaje on actor.nombre = personaje.a_nombre
inner join pelicula on pelicula.nombre = personaje.p_nombre
where actor.nombre ilike '%nicholson, jack%'
order by pelicula.calificacion desc;
```

## Pregunta 4
Los nombres de los personajes interpretados por mujeres, en películas de los 90s, con calificación mayor a 8,7.

```
%%sql
select personaje
from actor inner join personaje on actor.nombre = personaje.a_nombre
inner join pelicula on pelicula.nombre = personaje.p_nombre
where genero ilike '%F%'
and pelicula.calificacion > 8.7
and pelicula.anho between 1990 and 1999;
```

## Pregunta 5
Las películas de la saga "The Lord of the Rings" (usando el prefijo de su nombre) ordenadas por calificación y por año.

```
%%sql
select *
from pelicula
where nombre ilike '%lord of the rings%'
order by calificacion desc, anho asc;
```

## Pregunta 6
Los nombres de los actores que interpretan más de un personaje en la misma película.

```
%%sql
select actor.nombre, personaje.p_nombre, count(personaje)
from actor inner join personaje on actor.nombre = personaje.a_nombre
group by actor.nombre, personaje.p_nombre
having count(personaje) > 1
```

## Pregunta 7
Las películas en que actúan juntos Uma Thurman y Samuel L. Jackson.

```
%%sql
select p_nombre
from personaje
where a_nombre ilike '%Thurman, Uma%'
and p_nombre in (
    select p_nombre
    from personaje
    where a_nombre ilike '%Jackson, Samuel%');
```

## Pregunta 8
Las películas en que actúa Uma Thurman y **no** Samuel L. Jackson.

```
%%sql
select p_nombre
from personaje
where a_nombre ilike '%Thurman, Uma%'
and p_nombre not in (
    select p_nombre
    from personaje
    where a_nombre ilike '%Jackson, Samuel%');
```

## Pregunta 9
Los pares de actores que aparecen juntos en más de una película. Cada par debe aparecer una sola vez, es decir, si $(A,B)$ aparece, no debe aparecer $(B,A)$, pues es el mismo par. Tampoco se deben incluir pares de la forma $(A,A)$.

```
%%sql
select a1.nombre, a2.nombre
from actor a1 inner join actor a2 on a1.nombre < a2.nombre
where exists (
    select *
    from personaje p1 inner join personaje p2 on p1.p_nombre = p2.p_nombre
    where p1.a_nombre = a1.nombre and p2.a_nombre = a2.nombre)
group by a1.nombre, a2.nombre
having count(*) > 1;
```

## Pregunta 10
La película con calificación más alta.

```
%%sql
select nombre
from pelicula
where calificacion = (
    select max(calificacion)
    from pelicula);
```
